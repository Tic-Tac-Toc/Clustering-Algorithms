```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_datas = pd.read_csv('data.csv')
X = df_datas.drop(df_datas.columns[0], axis = 1).values
```

Try to display the data.

```python
time = df_datas.columns[1:]
plt.rcParams["figure.figsize"] = (20,10)
for elm in X:
    plt.plot(time, elm)

plt.legend(df_datas["StockName"])
plt.xlabel("Date")
plt.ylabel("Stock")
plt.show()
```

KMeans algorithm & print the sum squared error

```python
kmeans = KMeans(init = 'random', n_clusters=8, max_iter=1000, n_init=100).fit(X)
```

```python
print("SSE for {} clusters : {}".format(len(kmeans.cluster_centers_), kmeans.inertia_))
```

Better display for clusters

```python
cluster_map = pd.DataFrame()
cluster_map['cluster'] = kmeans.labels_
cluster_map['StockName'] = df_datas["StockName"]
```

```python
dic_cluster = {}

for i in range(len(kmeans.cluster_centers_)):
    dic_cluster[i] = cluster_map[cluster_map.cluster == i].StockName.values

for elm in dic_cluster:
    txt = "Cluster " + str(elm + 1) + " - " + ",".join(dic_cluster[elm])
    print(txt)
```
